```python
from ultralytics import YOLO
from sklearn.metrics import classification_report
import os
from pathlib import Path

model = YOLO(r"runs\classify\ddr-224-yolo26-nano2\weights\best.pt")

test_path = Path(dataset.location) / "test"

y_true = []
y_pred = []

# Loop over class folders
for class_id, class_name in enumerate(sorted(os.listdir(test_path))):
    class_folder = test_path / class_name
    
    for img_path in class_folder.glob("*.*"):
        result = model(img_path, verbose=False)[0]
        
        pred_class = result.probs.top1  # predicted index
        
        y_pred.append(pred_class)
        y_true.append(class_id)

# Print metrics
print(classification_report(y_true, y_pred, digits=4))
```
